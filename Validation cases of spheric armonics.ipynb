```python
import pandas as pd
import numpy as np
from scipy.special import sph_harm
from rwhist import compute_sph_coeffs, print_coeffs
```

```python
class SphereSectors():
    def __init__(self, num_theta, num_phi):
      # mesh in theta-phi
      self.theta_edges = np.linspace(0, np.pi, num_theta + 1)  # incluye polos
      self.phi_edges = np.linspace(0, 2 * np.pi, num_phi + 1)  # incluye cierre

      # coordinates of the spherical sector center
      self.theta = 0.5 * (self.theta_edges[:-1] + self.theta_edges[1:])
      self.phi = 0.5 * (self.phi_edges[:-1] + self.phi_edges[1:])

      self.mtheta, self.mphi = np.meshgrid(self.theta, self.phi, indexing='ij')       

      return

sectors = SphereSectors( num_theta = 50, num_phi = 50)
```

```python
num_theta=50
num_phi=50

theta_edges = sectors.theta_edges
phi_edges   = sectors.phi_edges

theta_center = sectors.theta
phi_center   = sectors.phi

fs = np.zeros((num_theta, num_phi), dtype=np.complex128)
Lmax=6
```

## Case 1: Uniform distribution

```python
fs[:, :] = 1.0 / (4 * np.pi) + 0.0j
coeffs = compute_sph_coeffs(fs, theta_edges, phi_edges, Lmax)
print("\n=== CASE 1: UNIFORM DISTRIBUTION ===")
print_coeffs(coeffs)

print("\n" + "="*80 + "\n")
```

## CASE 2: Dipole distribution

```python
for i, theta in enumerate(theta_center):
        for j, phi in enumerate(phi_center):
            fs[i, j] = (3 / (4 * np.pi)) * (np.cos(theta) ** 2) + 0.0j

coeffs = compute_sph_coeffs(fs, theta_edges, phi_edges, Lmax)
print("\n=== CASE 2: DIPOLE DISTRIBUTION ===")
print_coeffs(coeffs)
```

## CASE 3A: φ-constant distribution

La función wl = ∑ hist * S, que vale 0. Por lo tanto al dividir por el peso alm/wl sale un armónico infinito.

```python
for i, theta in enumerate(theta_center):
        fs[i, :] = (3 / (8 * np.pi)) * np.sin(3.*theta) + 0.0j

coeffs = compute_sph_coeffs(fs, theta_edges, phi_edges, Lmax)
print("\n=== CASE 3A: PHI-CONSTANT DISTRIBUTION ===")
print_coeffs(coeffs)
```

## CASE 3B: φ-symmetric distribution

```python
for i, theta in enumerate(theta_center):
        for j, phi in enumerate(phi_center):
            fs[i, j] = (3 / (8 * np.pi)) * np.sin(3.*theta) * np.abs(np.cos(2.*phi)) + 0.0j

coeffs = compute_sph_coeffs(fs, theta_edges, phi_edges, Lmax)
print("\n=== CASE 3B: PHI-SYMMETRIC DISTRIBUTION ===")
print_coeffs(coeffs)
```

## CASE 4: Equatorial symmetric distribution

```python
for i, theta in enumerate(theta_center):
        fs[i, :] = (3 / (16 * np.pi)) * np.sin(theta)**2 + 0.0j

coeffs = compute_sph_coeffs(fs, theta_edges, phi_edges, Lmax)
print("\n=== CASE 4: EQUATORIAL SYMMETRIC DISTRIBUTION ===")
print_coeffs(coeffs)
```

## CASE 5: Custom distribution 

```python
case = [
    {'l': 0, 'm': 0, 'coeff': 1.0 + 0.0j},
    {'l': 1, 'm': 0, 'coeff': 0.5 + 0.0j},
    {'l': 2, 'm': 1, 'coeff': 0.2 + 0.0j},
    {'l': 3, 'm': 2, 'coeff': 0.1 + 0.0j},
]

# Inicializamos la función angular
fs[:, :] = 0.0 + 0.0j

# Construcción correcta de la distribución
for c in case:
    l = c['l']
    m = c['m']
    coeff = c['coeff']

    for i, theta in enumerate(theta_center):
        for j, phi in enumerate(phi_center):
            # IMPORTANTE: orden correcto (phi, theta)
            Y_lm = sph_harm(m, l, phi, theta)
            fs[i, j] += coeff * Y_lm

# Cálculo de coeficientes
coeffs = compute_sph_coeffs(fs, theta_edges, phi_edges, Lmax)

print("\n=== CASE 5: CUSTOM DISTRIBUTION ===")
print_coeffs(coeffs)
```

## CASE 6: RANDOM DISTRIBUTION

```python
fs[:, :] = np.random.random_sample( size=(num_theta, num_phi) ) + np.random.random_sample( size=(num_theta, num_phi) ) * 1.0j

coeffs = compute_sph_coeffs(fs, theta_edges, phi_edges, Lmax)
print("\n=== CASE 6: RANDOM DISTRIBUTION ===")
print_coeffs(coeffs)
```
